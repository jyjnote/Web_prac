```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

url = "https://www.google.com/maps"
search_term = "교촌치킨"

# Chrome WebDriver 초기화
driver = webdriver.Chrome()
driver.get(url)

# 검색어 입력 및 검색 실행
search_box = driver.find_element(By.ID, "searchboxinput")
search_box.send_keys(search_term)
search_box.send_keys(Keys.ENTER)

# 검색 결과 로드를 위한 대기
time.sleep(5)

# XPath를 사용하여 스크롤바 요소 찾기
scrollbar_xpath = "//*[@id='QA0Szd']/div/div/div[1]/div[2]/div/div[1]/div/div/div[1]/div[1]"
scrollbar = driver.find_element(By.XPATH, scrollbar_xpath)

# 스크롤바 요소가 보이도록 스크롤
driver.execute_script("arguments[0].scrollIntoView(true);", scrollbar)

# 지정된 횟수만큼 스크롤 수행
for _ in range(10):
    driver.execute_script("arguments[0].scrollBy(0, 1000);", scrollbar)
    time.sleep(1)  # 스크롤 사이에 로딩 대기

# 페이지 소스를 BeautifulSoup 객체로 파싱
soup = BeautifulSoup(driver.page_source, 'html.parser')

# 검색 결과 링크 추출
store_links = soup.find_all("a", class_="hfpxzc")

for link in store_links:
    store_url = link['href']
    # 각 링크에 대한 상세 페이지 방문
    driver.get(store_url)
    time.sleep(5)  # 상세 페이지 로드 대기

    # 상세 페이지의 내용 파싱
    detailed_soup = BeautifulSoup(driver.page_source, 'html.parser')
    name_tag = detailed_soup.find("h1", class_="DUwDvf lfPIob")
    address_tag = detailed_soup.find("div", class_="Io6YTe fontBodyMedium kR99db")
    rating_tag = detailed_soup.find("div", class_="fontDisplayLarge")

    # 가게 이름, 주소, 평점 추출
    name = name_tag.get_text().strip() if name_tag else "이름 없음"
    address = address_tag.get_text().strip() if address_tag else "주소 없음"
    score = rating_tag.get_text().strip() if rating_tag else "평점 없음"

    # 결과 출력
    print(f"가게 이름: {name}, 주소: {address}, 평점: {score}")

# WebDriver 종료
driver.quit()
```
